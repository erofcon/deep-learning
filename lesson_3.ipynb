# Введение в нейронное прогнозирование: прямое распространение

### Простая нейронная сеть, делающая прогноз

```python
weight = 0.1

def neural_network(input, weight):
    prediction = input * weight
    return prediction
```

```python
number_of_toes = [8.5, 9.5, 10, 9]
input = number_of_toes[0]
pred = neural_network(input, weight)
print(pred)
```

### Прогнозирование с несколькими входами

```python
def w_sum(a,b):
    assert(len(a) == len(b))
    output = 0

    for i in range(len(a)):
        output += a[i] * b[i]

    return output

def neural_network(input, weights):
    pred = w_sum(input,weights)
    return pred
```

```python
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

input = [toes[0],wlrec[0],nfans[0]]
weights = [0.1, 0.2, 0]

pred = neural_network(input,weights)

print(pred)
```

Код с использованием numpy

```python
import numpy as np

weights = np.array([0.1, 0.2, 0])

def neural_network(input, weights):
    prediction = input.dot(weights)
    return prediction

toes = np.array([8.5, 9.5, 9.9, 9.0])
wlrec = np.array([0.65, 0.8, 0.8, 0.9])
nfans = np.array([1.2, 1.3, 0.5, 1.0])

input = np.array([toes[0],wlrec[0],nfans[0]])

pred = neural_network(input,weights)

print(pred)
```

### Прогнозирование с несколькими выходами

```python
def ele_mul(number,vector):
    output = [0,0,0]
    assert(len(output) == len(vector))

    for i in range(len(vector)):
        output[i] = number * vector[i]
        
    return output
```

```python
weights = [0.3, 0.2, 0.9]

def neural_network(input, weights):
    pred = ele_mul(input,weights)
    return pred
```

```python
wlrec = [0.65, 0.8, 0.8, 0.9]

input = wlrec[0]

pred = neural_network(input,weights)

print(pred)
```

### Прогнозирование с несколькими входами и выходами

```python
weights = [ [0.1, 0.1, -0.3], 
            [0.1, 0.2, 0.0],
            [0.0, 1.3, 0.1] ]


def w_sum(a,b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i] * b[i])
    return output

def vect_mat_mul(vect,matrix):
    assert(len(vect) == len(matrix))
    output = [0,0,0]
    for i in range(len(vect)):
        output[i] = w_sum(vect,matrix[i])
    return output

def neural_network(input, weights):
    pred = vect_mat_mul(input,weights)
    return pred
```

```python
toes =  [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65,0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

# Input corresponds to every entry
# for the first game of the season.

input = [toes[0],wlrec[0],nfans[0]]
pred = neural_network(input,weights)

print(pred)
```

### Прогнозирование на основе прогнозов

```python
ih_wgt = [ [0.1, 0.2, -0.1],
           [-0.1,0.1, 0.9],
           [0.1, 0.4, 0.1] ]

         
hp_wgt = [ [0.3, 1.1, -0.3],
           [0.1, 0.2, 0.0],
           [0.0, 1.3, 0.1] ]

weights = [ih_wgt, hp_wgt]

def neural_network(input, weights):
    hid = vect_mat_mul(input,weights[0])
    pred = vect_mat_mul(hid,weights[1])
    return pred

toes =  [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65,0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

input = [toes[0],wlrec[0],nfans[0]]
pred = neural_network(input,weights)

print(pred)
```

Версия с numpy

```python
import numpy as np


ih_wgt = np.array([ 
            [0.1, 0.2, -0.1],
            [-0.1,0.1, 0.9],
            [0.1, 0.4, 0.1]]).T



hp_wgt = np.array([  
            [0.3, 1.1, -0.3], 
            [0.1, 0.2, 0.0], 
            [0.0, 1.3, 0.1] ]).T

weights = [ih_wgt, hp_wgt]

def neural_network(input, weights):

    hid = input.dot(weights[0])
    pred = hid.dot(weights[1])
    return pred


toes =  np.array([8.5, 9.5, 9.9, 9.0])
wlrec = np.array([0.65,0.8, 0.8, 0.9])
nfans = np.array([1.2, 1.3, 0.5, 1.0])

input = np.array([toes[0],wlrec[0],nfans[0]])

pred = neural_network(input,weights)
print(pred)
```
